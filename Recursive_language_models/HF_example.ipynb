```python
from datasets import load_dataset

# Load the dataset with the 'dnd' config
dataset = load_dataset("oolongbench/oolong-real", "dnd", split="test")  # you can also use 'validation'

# Read the first record
record = dataset[0]

# Extract 'context_window_text' column
context_text = record.get("context_window_text", "")

# Write it to a file
with open("file.txt", "w", encoding="utf-8") as f:
    f.write(context_text)

# Print question and answer
question = record.get("question", "")
answer = record.get("answer", "")

print("Question:", question)
print("Answer:", answer)
```
